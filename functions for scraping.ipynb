### function to request HTML from various websites

```python
def get_soup(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    print(soup.prettify())
    return
```

```python
get_soup('https://www.saturn.de/de/campaign/angebote-computer-tablet')
```

```python
def get_soup(url):
    r = requests.get(url)
    soup = BeautifulSoup(r.text, 'html.parser')
    all_infos = soup.find_all('div', class_= 'ProductHeader__StyledHeadingWrapper-cwyxax-0 gQJaMt' )
    
    for info in all_infos:
        headline = info.find(class_ = 'Typostyled__StyledInfoTypo-sc-1jga2g7-0 clXCSd').text
        print(headline)
```

```python
get_soup('https://www.saturn.de/de/campaign/angebote-computer-tablet')
```
